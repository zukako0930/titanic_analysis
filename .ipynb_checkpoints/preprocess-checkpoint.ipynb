## タイタニックのほーしん
- Ageの補完をName,Fareを使って予測 -> Mr,Mrs,Missなどの平均
- とりあえず手法をいろいろやって見る

- PassengerID:　乗客ID
- Survived: 　　生存結果 (1: 生存, 2: 死亡)　
- Pclass: 　　　乗客の階級 1が一番位が高いそう
- Name: 　　 　乗客の名前
- Sex: 　　　　 性別
- Age: 　　　　 年齢
- SibSp 　　 　 兄弟、配偶者の数。
- Parch 　　　　両親、子供の数。
- Ticket 　　　 チケット番号。
- Fare 　　　　 乗船料金。
- Cabin 　　　　部屋番号
- Embarked 　　 乗船した港　Cherbourg、Queenstown、Southamptonの３種類があります

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

```python
df.Sex = df.Sex.map({'male':0,'female':1})
```

```python
dum_df = pd.concat((df,pd.get_dummies(df.Embarked)),axis=1)
```

```python
dum_df.loc[:,['Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','S','C','Q']]
```

```python
dum_df[dum_df.Pclass == 1].Age = dum_df.Age.fillna(dum_df[dum_df.Pclass == 1].Age.mean())
dum_df[dum_df.Pclass == 2].Age = dum_df.Age.fillna(dum_df[dum_df.Pclass == 2].Age.mean())
dum_df[dum_df.Pclass == 3].Age = dum_df.Age.fillna(dum_df[dum_df.Pclass == 3].Age.mean())
```

```python
dum_df.head(10)
```

```python
class1_df = dum_df[dum_df.Pclass == 1]
class2_df = dum_df[dum_df.Pclass == 2]
class3_df = dum_df[dum_df.Pclass == 3]
```

```python
class1_df['Age'] = class1_df['Age'].fillna(class1_df['Age'].mean())
class2_df['Age'] = class2_df['Age'].fillna(class2_df['Age'].mean())
class3_df['Age'] = class3_df['Age'].fillna(class3_df['Age'].mean())
```

```python
fill_age_df = pd.concat((class1_df,class2_df,class3_df),axis=0).sort_values(by=['PassengerId'],ascending=True)
```

```python
fill_age_df.isnull().sum()
```

```python
X = fill_age_df.loc[:,["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare",'S','C','Q']].values
y = fill_age_df.loc[:,["Survived"]].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0) #コンペはseed固定してなんぼ。
```

```python
param_grid = [
    {'C': [0.01, 0.1, 1], 'kernel': ['linear','rbf']}
]
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

# グリッドサーチのインスタンスを生成
gs_svc = GridSearchCV(SVC(),param_grid,cv=skf)
```

```python
# 教師データの型を修正(二重リストから一重のリストへ変換)
y_train = y_train.reshape((1,-1))[0]
y_test = y_test.reshape((1,-1))[0]
```

```python
# fitさせる(実行には少し時間がかかります)
gs_svc.fit(X_train,y_train)
```

```python
gs_svc.best_score_
```

```python
pd.DataFrame(X).isnull().sum()
```

```python
#ロジスティック回帰
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
# データの分割方法を指定
from sklearn.model_selection import KFold
kf_3 = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lr,X_train,y_train,cv=kf_3).mean()
```

```python
# ランダムフォレストによる学習
from sklearn.ensemble import RandomForestClassifier
rfc_1 = RandomForestClassifier(random_state=0, n_estimators=10)
rfc_1.fit(X_train, y_train)
```

```python
rfc_1.score(X_test,y_test)
```
